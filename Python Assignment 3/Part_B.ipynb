```python
# Import libraries 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

We consider numerical simulations of the stochastic differential equation for Geometric Brownian Motion given by 

$$
\dot S(t) = \mu S(t) + \sigma S(t) \xi(t)
$$

We consider the following parameter values:

```python
# Problem setup. 
# Set model and numerical parameters, and the initial condition.
tf = 5
Nsteps = 365 * tf
Npaths = 2000
S0 = 100
mu = 0.05
sigma = 0.3

# Generate the time grid and solution array
t, dt = np.linspace(0,tf,Nsteps+1,retstep=True)
S = np.zeros((Nsteps+1,Npaths))
root_dt = np.sqrt(dt)
  
# Time step starting from initial condition
S[0,:] = S0;
for n in range(Nsteps):
    F_of_S = mu * S[n,:]
    S[n+1,:] =  S[n,:] + dt * F_of_S + sigma * root_dt * S[n,:] * np.random.randn(Npaths)
    
# Plot paths
plt.plot(t,S)
plt.xlabel("t", fontsize=14)
plt.ylabel("S", fontsize=14)
plt.title("Sample paths", fontsize=14)
plt.show()
```

```python
# Compute the mean and standard deviation as function of time
S_mean = np.mean(S,1)
S_std = np.std(S,1)

# Plot shaded region between S_mean - S_std and S_mean + S_std 
plt.fill_between(t, 
                 S_mean - S_std, 
                 S_mean + S_std, 
                 alpha=0.5, color="darkorange")

# Plot the mean itself
plt.plot(t,S_mean, linewidth=2)

# labels
plt.xlabel("t", fontsize=18)
plt.ylabel("S", fontsize=18)
plt.title("Mean +/- Std.", fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
# Plot histogram of all solutions at the final time. 
plt.hist(S[-1,:], bins=50, density=True)
plt.xlim(0, 400)
plt.xticks(np.arange(0, 450, 50))
plt.xlabel("S", fontsize=14)
plt.ylabel("density", fontsize=14)
plt.title("histogram at final time", fontsize=14)
plt.show()
```
